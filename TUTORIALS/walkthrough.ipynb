```python
import socket
import os
import sys
```

```python
IP = "127.0.0.1"
PORT = 12345
```

```python
def stream_client():
    
    try:
        with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as sockfd:
            print(f"client: attempting to connect to server at {IP}:{PORT}")
            sockfd.connect((IP, PORT))
            print("client: connection established")
            response = sockfd.recv(4096)
            print(f"client: received '{response.decode()}'\n")
            sockfd.close()

    except OSError as e:
        print("socket error:", e)
        sockfd.close()
        sys.exit(1)
    
```

```python
stream_client()
```

```python
def stream_client_2(msg:str):
    try:
        with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as sockfd:
            print(f"client: attempting to connect to server at {IP}:{PORT}")
            sockfd.connect((IP, PORT))
            print("client: connection established")
            msg = msg.encode()
            sockfd.send(msg)

            response = sockfd.recv(4096)
            if response.decode() == "butt poop":
                print("LOL")
            else:
                print(response.decode())
            sockfd.close()

    except OSError as e:
        print("socket error:", e)
        sockfd.close()
        sys.exit(1)

stream_client_2("poop buttt")
```

```python
def datagram_client():
    try:
        with socket.socket(socket.AF_INET, socket.SOCK_DGRAM) as sockfd:
            print(f"client: sending datagram to {IP}:{PORT}")
            msg = b"Hello, Datagram Server!"
            bytes_sent = sockfd.sendto(msg, (IP, PORT))

            print(f"client: sent {bytes_sent} bytes to {IP}:{PORT}")
            sockfd.close()

    except OSError as e:
        print("socket error:", e)
        sockfd.close()
        sys.exit(1)
```

```python
datagram_client()
```

As a sanity check, modify both clients and their corresponding servers so they exchange messages back and forth (send → receive → send → receive). Good luck!
